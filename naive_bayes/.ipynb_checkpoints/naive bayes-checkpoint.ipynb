```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
```

```python
train_bow_df = pd.read_csv("../data/train_bow.csv")
train_tfidf_df = pd.read_csv("../data/train_tfidf.csv")
test_bow_df = pd.read_csv("../data/test_bow.csv")
test_tfidf_df = pd.read_csv("../data/test_tfidf.csv")
```

```python
X_train_bow = train_bow_df.drop("isFakeNews", axis = 1)
y_train_bow = train_bow_df["isFakeNews"]
X_test_bow = test_bow_df.drop("isFakeNews", axis = 1)
y_test_bow = test_bow_df["isFakeNews"]
```

```python
X_train_tfidf = train_tfidf_df.drop("isFakeNews", axis = 1)
y_train_tfidf = train_tfidf_df["isFakeNews"]
X_test_tfidf = test_tfidf_df.drop("isFakeNews", axis = 1)
y_test_tfidf = test_tfidf_df["isFakeNews"]
```

# Gaussian Naive Bayes

```python
# Set Model
model = GaussianNB()

# BOW
model.fit(X_train_bow, y_train_bow)
bow_pred = model.predict(X_test_bow)
print("BOW Accuracy: {}".format(accuracy_score(y_test_bow, bow_pred)))

# TFIDF
model.fit(X_train_tfidf, y_train_tfidf)
bow_pred = model.predict(X_test_tfidf)
print("TFIDF Accuracy: {}".format(accuracy_score(y_test_tfidf, bow_pred)))
```

# Multinomial Naive Bayes

```python
bow_accuracy_mnb = dict()
tfidf_accuracy_mnb = dict()

for alpha_val in range(1, 101):
    alpha_val = alpha_val / 100
    
    # Set Model
    model = MultinomialNB(alpha = alpha_val)

    # BOW
    model.fit(X_train_bow, y_train_bow)
    bow_pred = model.predict(X_test_bow)
    bow_accuracy_mnb[alpha_val] = accuracy_score(y_test_bow, bow_pred)

    # TFIDF
    model.fit(X_train_tfidf, y_train_tfidf)
    bow_pred = model.predict(X_test_tfidf)
    tfidf_accuracy_mnb[alpha_val] = accuracy_score(y_test_tfidf, bow_pred)
```

```python
bow_accuracy_mnb_uniform_prior = dict()
tfidf_accuracy_mnb_uniform_prior = dict()

for alpha_val in range(1, 101):
    alpha_val = alpha_val / 100
    
    # Set Model
    model = MultinomialNB(alpha = alpha_val, fit_prior = False)

    # BOW
    model.fit(X_train_bow, y_train_bow)
    bow_pred = model.predict(X_test_bow)
    bow_accuracy_mnb_uniform_prior[alpha_val] = accuracy_score(y_test_bow, bow_pred)

    # TFIDF
    model.fit(X_train_tfidf, y_train_tfidf)
    bow_pred = model.predict(X_test_tfidf)
    tfidf_accuracy_mnb_uniform_prior[alpha_val] = accuracy_score(y_test_tfidf, bow_pred)
```

# Bernoulli Naive Bayes

```python
bow_accuracy_bnb = dict()
tfidf_accuracy_bnb = dict()

for alpha_val in range(1, 101):
    alpha_val = alpha_val / 100
    
    # Set Model
    model = BernoulliNB(alpha = alpha_val)

    # BOW
    model.fit(X_train_bow, y_train_bow)
    bow_pred = model.predict(X_test_bow)
    bow_accuracy_bnb[alpha_val] = accuracy_score(y_test_bow, bow_pred)

    # TFIDF
    model.fit(X_train_tfidf, y_train_tfidf)
    bow_pred = model.predict(X_test_tfidf)
    tfidf_accuracy_bnb[alpha_val] = accuracy_score(y_test_tfidf, bow_pred)
```

```python
bow_accuracy_bnb_uniform_prior = dict()
tfidf_accuracy_bnb_uniform_prior = dict()

for alpha_val in range(1, 101):
    alpha_val = alpha_val / 100
    
    # Set Model
    model = BernoulliNB(alpha = alpha_val, fit_prior = False)

    # BOW
    model.fit(X_train_bow, y_train_bow)
    bow_pred = model.predict(X_test_bow)
    bow_accuracy_bnb_uniform_prior[alpha_val] = accuracy_score(y_test_bow, bow_pred)

    # TFIDF
    model.fit(X_train_tfidf, y_train_tfidf)
    bow_pred = model.predict(X_test_tfidf)
    tfidf_accuracy_bnb_uniform_prior[alpha_val] = accuracy_score(y_test_tfidf, bow_pred)
```

```python
bow_acc_df = pd.DataFrame.from_dict(bow_accuracy_mnb, orient = "index", columns = ["mnb"])
bow_acc_df["mnb_uniform_prior"] = bow_accuracy_mnb_uniform_prior.values()
bow_acc_df["bnb"] = bow_accuracy_bnb.values()
bow_acc_df["bnb_uniform_prior"] = bow_accuracy_bnb_uniform_prior.values()
sns.lineplot(data = bow_acc_df)
```

```python
tfidf_acc_df = pd.DataFrame.from_dict(tfidf_accuracy_mnb, orient = "index", columns = ["mnb"])
tfidf_acc_df["mnb_uniform_prior"] = tfidf_accuracy_mnb_uniform_prior.values()
tfidf_acc_df["bnb"] = tfidf_accuracy_bnb.values()
tfidf_acc_df["bnb_uniform_prior"] = tfidf_accuracy_bnb_uniform_prior.values()
sns.lineplot(data = tfidf_acc_df)
```

```python
max_bow_key = max(bow_accuracy_bnb, key = bow_accuracy_bnb.get)
print(max_bow_key, bow_accuracy_bnb[max_bow_key])
max_tfidf_key = max(tfidf_accuracy_bnb, key = tfidf_accuracy_bnb.get)
print(max_tfidf_key, tfidf_accuracy_bnb[max_tfidf_key])
```
